# Import all required LibrariesÂ¶

```python
import pandas
import numpy
```

# Read data

```python
data = pandas.read_csv('Titanic.csv')
data.head(10)
```

# Drop Unnessesary columns from the data

```python
data = data.drop(columns = ['PassengerId', 'Name', 'Ticket'])
data.head(10)
```

# Replacing non numerical values to numerical values

```python
from sklearn.preprocessing import LabelEncoder
le_sex = LabelEncoder()
data['sex'] = le_sex.fit_transform(data['Sex'])
data.head(10)
```

```python
data = data.drop(columns = ['Sex'])
data.head(10)
```

# Filling null values with the mean of remaining values in the column

```python
data.Age = data.Age.fillna(data.Age.mean())
data.head(10)
```

# Dividing the data into dependent and independent variables

```python
Y = data.Survived
X = data.drop(columns = ['Survived'])
X, Y
```

# Splitting data into training data and testing data 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
```

# Standardize of a dataset

https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
X_train = SS.fit_transform(X_train)
X_test = SS.fit_transform(X_test)
X_train, X_test
```

# Import DesicionTreeClassifier from the sklearn library

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
```

# Training model by providing the training data into the model

```python
model.fit(X_train, y_train)
```

# Training model by providing the training data into the model

```python
y_pred = model.predict(X_test)
y_pred
```

# Printing metrics of the model

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)
acc
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm
```
